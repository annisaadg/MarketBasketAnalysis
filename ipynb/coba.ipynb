```python
data = pd.read_csv("../datasets/ind.csv", sep = ";")
```

```python
data1 = data[["BillNo", "Itemname", "Quantity"]].copy()
```

```python
data1.head()
```

```python
#mengubah dari 0,0 menjadi 0
def one_hot_encoding(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
data1["Quantity"] = data1["Quantity"].apply(one_hot_encoding)
```
