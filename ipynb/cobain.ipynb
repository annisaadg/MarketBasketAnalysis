```python
import numpy as np
import pandas as pd
from pyECLAT import ECLAT
```

```python
data = pd.read_csv("output1.csv",on_bad_lines='skip',header=None)
```

```python
dataEclat = data
dataEclat.head(100)
```

```python
eclat_instance = ECLAT(data=dataEclat, verbose=True)
```

```python
eclat_instance.df_bin
```

```python
get_ECLAT_indexes, get_ECLAT_supports = eclat_instance.fit(min_support=0.02,
                                                           min_combination=1,
                                                           max_combination=3,
                                                           separator=' & ',
                                                           verbose=True)
```
