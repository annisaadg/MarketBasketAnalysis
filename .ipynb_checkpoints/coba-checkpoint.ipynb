```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#data = pd.read_excel("Online Retail.xlsx")
#Membaca Dataset

#Meload Dataset dari file csv
data = pd.read_csv("ind.csv", sep = ";")

#Menampilkan 20 data teratas
data.head(20)
```

```python
#cek null value
data.isnull().sum()
```

```python
#drop na value pada kolom Itemname
data = data.dropna(subset=["Itemname"])
```

```python
#cek null value
data.isnull().sum()
```

```python
#menghitung jumlah 
data["Itemname"].value_counts()
```

```python
#Menampilkan data dengan CustomerID Null
data[data['CustomerID'].isnull()]
```

```python
#Mengisi semua na value dengan 0
data = data.fillna(0)
```

```python
data[data["CustomerID"].isnull()]
```

```python
data.isnull().sum()
```

```python
#mengubah kedalam bentuk date
data["Date"] = pd.to_datetime(data["Date"])
```

```python
#Mengubah tanda pada price ke dalam koma
data["Price"] = data["Price"].str.replace(",",".")
data["Price"] = pd.to_numeric(data["Price"])
```

```python
data["Price"]
```

```python
data["Sales"] = data["Quantity"] * data["Price"]
```

```python
data.head()
```

```python
top20items = pd.DataFrame(data["Itemname"].value_counts().head(20))
top20items = top20items.reset_index()
top20items.columns = ["Itemname","Frequency"]
labels = top20items["Itemname"]
sizes = top20items["Frequency"]
top20items
```

```python
basket = data.groupby(["BillNo","Itemname"])["Quantity"].sum().unstack().reset_index().fillna(0).set_index("BillNo")
basket
```

```python
#mengubah dari 0,0 menjadi 0
def one_hot_encoding(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket = basket.applymap(one_hot_encoding)
```

```python
basket
```

```python
#Menemuka Frequent Itemsets dengan menggunakan algoritma apriori
frequent_itemsets = apriori(basket, min_support=0.02, use_colnames=True)
```

```python
#Menerapkan Association Rule 
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```

```python
rules = rules.sort_values("lift", ascending = False)
```
