```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_csv("ind.csv", sep = ";")
data.head(20)
```

```python
data = data.dropna(subset=["Itemname"])
```

```python
data = data.fillna(0)
```

```python
data["Date"] = pd.to_datetime(data["Date"])
```

```python
basket = data.groupby(["BillNo","Itemname"])["Quantity"].sum().unstack().reset_index().fillna(0).set_index("BillNo")
basket
```

```python
# Group items by BillNo and concatenate the Itemname values
grouped_items = data.groupby("BillNo")["Itemname"].apply(lambda x: ' '.join(x)).reset_index()

# Rename the column
grouped_items = grouped_items.rename(columns={"Itemname": "Item_List"})

# Show the result
print(grouped_items)
```

```python
basket2 = data.groupby("BillNo")["Itemname"].value_counts()
basket2.head(100)
```

```python
#mengubah dari 0,0 menjadi 0
def one_hot_encoding(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket = basket.applymap(one_hot_encoding)
basket
```

```python
#let us return items and ietmsets with atleast 5% support:
freq_items=apriori(basket,min_support=0.02,use_colnames=True)
freq_items
```

```python
rules = association_rules(freq_items, metric="lift", min_threshold=1)
```
